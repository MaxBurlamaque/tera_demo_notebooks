```python
def produtos_por_carrinho(carrinho):
    produto = df[df["carrinho"]==carrinho].produto.to_list()
    desc = df[df["carrinho"]==carrinho].descricao.to_list()
    return [produto, desc]
    
produtos = []
descs = []
for i in lista_carrinhos[0]:
    x = produtos_por_carrinho(i)
    produtos.append(x[0])
    descs.append(x[1])
    
for a in descs:
    print("\n\n----------------------")
    print(a)
```

```python
df_testes_carrinho['id_produtos'] = produtos
df_testes_carrinho['desc_produtos'] = descs
df_testes_carrinho
```

```python
df_testes_carrinho.to_csv("testes_carrinho.csv", index=False)
```

### Gerando base de dados exemplar para apresentação no streamlit

```python
pesquisa = pd.read_csv("dados_pesquisa1.csv", sep=',')
pesquisa = pesquisa.iloc[:10]
pesquisa.head(3)
```

```python
pesquisa.shape
```

```python
prods = pd.DataFrame(columns=['ID', "DESC"])

eval(pesquisa.iloc[1,1])

for index, row in pesquisa.iterrows():
    for i in range(len(eval(row['desc_produtos']))):
        prods.loc[-1] = [eval(row['id_produtos'])[i], eval(row['desc_produtos'])[i]]
        prods.index = prods.index + 1
        prods = prods.sort_index()
prods.shape
```

```python
prods.duplicated(subset='ID').sum()
```

```python
prods = prods.drop_duplicates(subset='ID')
```

```python
prods.shape
```

```python
prods.to_csv("produtos_short.csv", index=False)
```
